# Edge detector

## Sobel X

```
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

soble_x_image = cv2.filter2D(image, -1, sobel_x)
plt.imshow(soble_x_image)
```

## Sobel Y

```
sobel_y = np.array([[1, 2, 1],
                    [0, 0, 0],
                    [-1, -2, -1]])

soble_y_image = cv2.filter2D(image, -1, sobel_y)
plt.imshow(soble_y_image)
```

## Lapacian

```
lapacian_1 = np.array([ [0, 1, 0],
                        [1, -4, 1],
                        [0, 1, 0]])

lapacian_1_image = cv2.filter2D(image, -1, lapacian_1)
plt.imshow(lapacian_1_image)
```

```
lapacian_2 = np.array([ [-1, -1, -1],
                        [-1,  8, -1],
                        [-1, -1, -1]])

lapacian_2_image = cv2.filter2D(image, -1, lapacian_2)
plt.imshow(lapacian_2_image)
```

## Highboost Filtering

```
low_pass = np.ones((3,3))/9 # Average fillter

low_pass_image = cv2.filter2D(image, -1, low_pass)
fhp = image - low_pass_image
shapping_image = image + 1*fhp
boot_image = image + 3*fhp

f, axarr = plt.subplots(1,3,figsize=(15,15))
axarr[0].imshow(shapping_image)
axarr[1].imshow(boot_image)
axarr[2].imshow(image)
```
